LISTING ALL PARTICULAR SHEET IN THE EXCEL

```python
import pandas as pd

# Load the Excel file
file_path = "/content/World Population dataset.xlsx"  # Ensure correct filename
xls = pd.ExcelFile('/content/World Population dataset.xlsx', engine="openpyxl")  # Use engine="openpyxl" for .xlsx files

# Print available sheet names
print("Available sheets:", xls.sheet_names)

```

SELECTING A SPECIFIC SHEET

```python
# Load a specific sheet (replace 'Sheet1' with the actual sheet name)
df = pd.read_excel(file_path, sheet_name="Morocco", engine="openpyxl")
unwanted_columns = ['Unnamed: 3', 'Unnamed: 4',
       'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9',
       'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13',
       'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17',
       'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21']

df.drop(unwanted_columns, axis = 1, inplace=True)
# Display the first few rows
df.head()
```

**# ANALYSIS; AREA PLOT OF MOROCCO POPULATION SPREAD ACROSS AGE GROUPS OVER THE YEARS.**

```python
plt.figure(figsize=(10,6))

# Define a color palette for each age group
colors = { '0-14': 'green', '15-64': 'yellow', '65 and above': 'blue'}

# Create stacked area chart
for age_group in df["Age group"].unique():
    subset = df[df["Age group"] == age_group]
    plt.fill_between(subset["Year"], subset["Population"], color=colors[age_group], label=age_group, alpha=.7)

# Customize the plot
plt.title("Morocco Population Spread Over Time by Age Group")
plt.xlabel("Year")
plt.ylabel("Population (millions)")
plt.legend(title="Age Group")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
